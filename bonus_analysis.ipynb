```python
from sqlalchemy import create_engine
import seaborn as sns
import pandas as pd
from sql_statement import sql_query
import matplotlib.pyplot as plt
from config import (user, password)
```

```python
# create connection
engine = create_engine(f"postgresql://{user}:{password}@localhost:5432/sql_hw")
conn = engine.connect()
```

```python
# read_in data from query
employee_data = pd.read_sql(sql_query,conn)
employee_data.head()
```

```python
employee_data.info()
```

```python
plt.figure(figsize= (20,8))
sns.boxplot(x = employee_data["title"],y = employee_data["salary"])
plt.title("Salary Ranges")
plt.show()
```

```python
average_pay = (employee_data.groupby("title")
               .agg(average_salary = ("salary","mean"))
               .sort_values(by = "average_salary", ascending = False)
               .reset_index())

plt.figure(figsize = (15,10))
sns.barplot(x = "title",y = "average_salary", data = average_pay)
plt.show()
```
